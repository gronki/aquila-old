```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
with os.popen('echo fit/al1.fit | ../bin/aqfindstar','r') as f:
    l1 = np.loadtxt(f,skiprows=2)
with os.popen('echo fit/al2.fit | ../bin/aqfindstar','r') as f:
    l2 = np.loadtxt(f,skiprows=2)
```

```python
plt.plot(l1[:,1],l1[:,2],'ro')
plt.plot(l2[:,1],l2[:,2],'go')
plt.show()
```

```python
with open('temp/l.txt','w') as f:
    f.write('2145 1428\n')
    f.write('%d\n' % len(l1[:,1]))
    for l in l1:
        f.write(' %.2f %.2f %.2f \n' % ( l[0], l[1], l[2] ) )
    f.write('%d\n' % len(l2[:,1]))
    for l in l2:
        f.write(' %.2f %.2f %.2f \n' % ( l[0], l[1], l[2] ) )
```
